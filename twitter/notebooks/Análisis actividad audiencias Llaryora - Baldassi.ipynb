```python
import networkx as nx
from twitter_api import API_HANDLER as TW
import json
```

```python
candidatos = [
    'HectorBaldassi',
    'MartinLlaryora',
]
```

```python
cand_ids = {'HectorBaldassi': '845696348', 'MartinLlaryora':'325778405'}
```

```python
g = nx.read_graphml('red_cands5k.graphml')
```

```python
# de estos usuarios tenemos TLs
with open("tl_uids.json") as f:
    tl_uids = json.load(f)
len(tl_uids)
```

```python
# Combinamos tweets de los 3 d√≠as
```

```python
from collections import defaultdict
```

```python
tweets = defaultdict(list)
```

```python
for day in [12, 13, 14]:
    with open('tweets_07-%d.json' % day) as f:
        for uid, tws in json.load(f).items():
            tweets[uid] += tws
```

```python
len(tweets)
```

```python
seguidores = {c: [uid for uid in g.predecessors(cand_ids[c]) if uid in tl_uids] for c in candidatos}
```

```python
# No hace falta, no se siguen mutuamente
# seguidores['HectorBaldassi'].remove(cand_ids['MartinLlaryora'])
# seguidores['MartinLlaryora'].remove(cand_ids['HectorBaldassi'])
```

```python
len(seguidores['HectorBaldassi'])
```

```python
len(seguidores['MartinLlaryora'])
```

```python
# Agrupamos tweets
```

```python
tweets_seguidores_cands = {c: [] for c in candidatos}
```

```python
for c in candidatos:
    for uid in seguidores[c]:
        tweets_seguidores_cands[c] += tweets[uid]
```

```python
len(tweets_seguidores_cands['HectorBaldassi'])
```

```python
len(tweets_seguidores_cands['MartinLlaryora'])
```

```python
tweets_seguidores_counts = {c: [] for c in candidatos}
```

```python
for c in candidatos:
    for uid in seguidores[c]:
        tweets_seguidores_counts[c].append(len(tweets[uid]))
```

```python
from bokeh.charts import Histogram, output_notebook, show
output_notebook()
```

```python
# bins = list(np.arange(0.3,1.01,0.05))
histp = Histogram(tweets_seguidores_counts['HectorBaldassi'])
# , xlabel="Precision score", ylabel='# of users')

show(histp)
```

```python
seguidores_activos = {c: [] for c in candidatos}
```

```python
for c in candidatos:
    for count in tweets_seguidores_counts[c]:
        if count:
            seguidores_activos[c].append(count)
```

```python
len(seguidores_activos['MartinLlaryora'])
```

```python
len(seguidores_activos['HectorBaldassi'])
```

```python
bins = range(0,101,10)
histp = Histogram(seguidores_activos['MartinLlaryora'], bins=bins, xlabel="#tweets", ylabel='# usuarios')

show(histp)
```

```python
bins = range(0,151,10)
histp = Histogram(seguidores_activos['HectorBaldassi'], bins=bins, xlabel="#tweets", ylabel='# usuarios')

show(histp)
```
